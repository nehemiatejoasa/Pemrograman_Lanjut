```python
# UTS PEMROGRAMAN LANJUTAN NEHEMIA TEJO ASA (18117008) & DHEANTY MAHARANI (18116029)

import pandas as pd
import re
```

```python
url = 'https://github.com/eueung/pilrek/raw/master/pilrek.csv'
df  = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df[df['Harapan Anda'].isnull()]['Kategori Anda'].value_counts().plot(kind='line')
# Untuk yang tidak mengisi pada bagian Harapan Anda (NaN) isi Kategori Anda ada apa saja
```

```python
df[df['Kategori Umur Anda'] == '< 25 th']['CaRek Pilihan'].value_counts().plot(kind='bar')
# Untuk Kategori Umur Anda <25 th carek pilihannya siapa saja 
```

```python
df[df['Kategori Umur Anda'] == '25 - 40 th']['CaRek Pilihan'].value_counts().plot(kind='bar')
# Untuk Kategori Umur Anda 25 - 40 th carek pilihannya siapa saja 
```

```python
df[df['Kategori Umur Anda'] == '> 40 th']['CaRek Pilihan'].value_counts().plot(kind='bar')
# Untuk Kategori Umur Anda >40 th carek pilihannya siapa saja 
```
